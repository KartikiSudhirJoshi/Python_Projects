```python
#predict the house price
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/aman9801/multiple-linear-regression-on-housing-dataset/main/mlr_housing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
mainroad=LabelEncoder()
guestroom=LabelEncoder()
basement=LabelEncoder()
hotwaterheating=LabelEncoder()
prefarea=LabelEncoder()
furnishingstatus=LabelEncoder()
airconditioning=LabelEncoder()
```

```python
df['mainroad']=mainroad.fit_transform(df['mainroad'])
df['guestroom']=guestroom.fit_transform(df['guestroom'])
df['basement']=basement.fit_transform(df['basement'])
df['hotwaterheating']=hotwaterheating.fit_transform(df['hotwaterheating'])
df['prefarea']=prefarea.fit_transform(df['prefarea'])
df['furnishingstatus']=furnishingstatus.fit_transform(df['furnishingstatus'])
df['airconditioning']=airconditioning.fit_transform(df['airconditioning'])

df
```

```python
#independent and dependent variable
features_cols=['area'	,'bedrooms','bathrooms',	'stories',	'mainroad',	'guestroom'	,'basement','hotwaterheating'	,'airconditioning','parking','prefarea','furnishingstatus']
x=df[features_cols]
y=df.price
```

```python
x
```

```python
y
```

```python
# Split The Dataset Into Training And Testing Subset
from sklearn.model_selection import train_test_split
# test_size=0.2 means 20% of the data will be used for Testing and remaining 80% used to training
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print('Original Shape')
print(x.shape)
print(y.shape)
print('After Splitting')
print('Training set :',x_train.shape)
print(y_train.shape)
print('Testing set :',x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(x_train,y_train)
```

```python
pred_values = mlr.predict(x_test)
pred_values
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,pred_values)
print('Accuracy of model :',accuracy*100)
```
